```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn.metrics as metrics

from tensorflow import keras
from tensorflow.keras import layers

from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
# Define o número de classes e a forma recebida 28 x 28 
num_classes = 10
input_shape = (28, 28, 1)

#carrega os dados, são 60 mil imagens de 0 aconfusion_matrixa de cinza pra treino
#e 10 mil imagens para testes

(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()
```

```python
#confirmando a quantidade e forma dos dados recebidos 

print("Forma das amostrass de treino:", x_train.shape)
print(x_train.shape[0], "Amostras de treino")
print(x_test.shape[0], "Amostra de testes")
print(y_train.shape[0], "rótulos de treino")
print(y_test.shape[0], "rótulos de teste")
```

```python
#análise do tipo de imagem recebida

for img in range(10):
    plt.subplot(2,5,img+1)
    plt.imshow(x_train[img])
```

```python
# redefinindo as imagens para o formato(28, 28, 1)
x_train = np.expand_dims(x_train, -1)
x_test = np.expand_dims(x_test, -1)

# normalizando os dados de 0-255 para 0-1 
x_train = x_train.astype("float32") / 255
x_test = x_test.astype("float32") / 255

# convert class vectors to binary class matrices
y_train = keras.utils.to_categorical(y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)

print(x_train.shape)
```

```python
model = keras.Sequential( #modelo padrão de rede neural com camadas ocultas e camada de saída e pesos.
    [
        keras.Input(shape=input_shape),
        layers.Conv2D(32, kernel_size=(3, 3), input_shape = (28, 28, 1), activation="relu"),
        layers.MaxPooling2D(pool_size=(2, 2)),#Trata a imegame de diversos angulos
        layers.Conv2D(64, kernel_size=(3, 3),input_shape = (28, 28, 1), activation="relu"),
        layers.MaxPooling2D(pool_size=(2, 2)),
        layers.Flatten(), #tranforma a imagem 28x28 em 1D
        layers.Dropout(0.5),
        layers.Dense(num_classes, activation="softmax"), #camada de saída (no dense a camda se conecta com todas as seguintes)
    ]
)

model.summary()
```

```python
#compilação da rede

#adam modelo de ajuste de pesos gradiente descendente otimizado
#loss modelo de definição do erro 
#metrics sistema de avaliação do modelo - acurácia 
model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])

#finalmente o treinaamento da máquina
atch_size = 128 #de quantas e quantas amostras ele ajusta os pesos
epochs = 15 #quantas vezes executa o ajuste dos pesos

model.fit(x_train, y_train, epochs=epochs,batch_size = atch_size , validation_split=0.1)
        
```

```python
#plotando a matriz de confusão dos testes 10 0000 amostras

previsoes = model.predict(x_test)
previsoes=model.predict(x_test) 
previsoes=np.argmax(previsoes, axis=1)
y_test=np.argmax(y_test, axis=1)
cm = confusion_matrix(y_test, previsoes)

ax= plt.subplot()
sns.heatmap(cm, annot=True, ax = ax); #annot=True to annotate cells

# labels, title and ticks
ax.set_xlabel('Predicted labels');ax.set_ylabel('True labels'); 
ax.set_title('Confusion Matrix'); 
ax.xaxis.set_ticklabels(['predicted', 'true']); ax.yaxis.set_ticklabels(['true', 'predicted']);
```

```python
# Acurácia
precisao = accuracy_score(y_test,previsoes)
print(precisao)
```
